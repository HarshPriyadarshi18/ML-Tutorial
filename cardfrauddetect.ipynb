```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data_sets=pd.read_csv('creditcard.csv  ')
```

```python
data_sets.shape
```

```python
data_sets.head()
```

```python
data_sets.isnull().sum()
```

```python
data_sets['Class'].value_counts()
```

```python
#under sampling bcoz variation is too much high between classes
legit=data_sets[data_sets.Class==0]
fraud=data_sets[data_sets.Class==1]
print(legit.shape)
print(fraud.shape)  
```

```python
#random sampling
Legit_sample=legit.sample(n=492)
```

```python
#concatenating two dataframes
new_dataset=pd.concat([Legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
#model evaluation
#accuracy score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data : ',test_data_accuracy)
```
